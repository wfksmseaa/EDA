```python
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import seaborn as sns

plt.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read data
df = pd.read_csv("movies.csv")
```

```python
# Top 5 data
df.head()
```

```python
# shape of data
df.shape
```

```python
# check is there any null value and count
df.isna().sum()
```

```python
# remove the row that has null values
df.dropna(inplace= True)
```

```python
# data types for each columns
df.dtypes
```

```python
# change data type of columns
df['budget'] = df['budget'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df.dtypes
```

```python
# sort values from gross
df.sort_values(by=['gross'], ascending=False,inplace=True)
```

```python
# make the data able to scrolled
pd.set_option('display.max_rows', None)
df.head(15)
```

```python
# scatter plot budget vs gross
plt.scatter(x= df['budget'],y =df['gross'])

plt.title("Plot Budget Vs Gross Earnings")
plt.xlabel("Budget (Billon Dollar)")
plt.ylabel("Gross (Billion Dollar)")

plt.show()
```

```python
# regression between gross and budget using seaborn

sns.regplot(x='budget',y ='gross', data= df , scatter_kws={'color' : 'darkorange'}, line_kws={'color':'black'})

```

```python
# correlation between budget and gross
df.corr()
```

```python
# Correlation matrix
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Matrix")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# sorted correlation pairs
corr_pairs = df.corr().unstack()
corr_pairs
```

```python
# sorted corr pairs
corr_pairs.sort_values(ascending=False)
```

```python
# high correlation 
high_corr = corr_pairs[(corr_pairs > 0.5)]
```

```python
high_corr.sort_values()
```
